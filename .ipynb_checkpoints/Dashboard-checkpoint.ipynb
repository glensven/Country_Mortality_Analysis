```python
#Dashboard construction
dash_df = pd.DataFrame(
    [['BTC'], 
     ['ETH'], 
     ['Doge']], 
     columns=['Crypto']
)
dash_df

# Class containing dropdown menu for cryptos, relevant plots and dataframes
class CryptoDashboard(param.Parameterized):
    
    #Crypto dropdown selector
    Crypto = param.ObjectSelector(default='BTC', objects=list(dash_df.Crypto.unique()))
    
    def rel_data(self):
        crypto_data = dash_df[(dash_df.Crypto==self.Crypto)].copy()
        return crypto_data

    def logo_png(self):
        data= self.rel_data()
        for dict in logo_dict:
            for k,v in dict.items():
                if v == self.Crypto:
                    logo=dict["Logo"]
        return logo
    
    def heat_plot_d(self):
        data= self.rel_data()
        for dict in heat_dict_d:
            for k,v in dict.items():
                if v == self.Crypto:
                    hmap=dict["Heat_d"]
        return hmap
    
    def heat_plot_w(self):
        data= self.rel_data()
        for dict in heat_dict_w:
            for k,v in dict.items():
                if v == self.Crypto:
                    hmap=dict["Heat_w"]
        return hmap
    
    def heat_plot_m(self):
        data= self.rel_data()
        for dict in heat_dict_m:
            for k,v in dict.items():
                if v == self.Crypto:
                    hmap=dict["Heat_m"]
        return hmap
    
    def heat_plot_y(self):
        data= self.rel_data()
        for dict in heat_dict_y:
            for k,v in dict.items():
                if v == self.Crypto:
                    hmap=dict["Heat_y"]
        return hmap
    
    def line_plot_d(self):
        data= self.rel_data()
        for dict in lines_dict_d:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Line_d"]
        return line
    
    def line_plot_w(self):
        data= self.rel_data()
        for dict in lines_dict_w:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Line_w"]
        return line
    
    def line_plot_m(self):
        data= self.rel_data()
        for dict in lines_dict_m:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Line_m"]
        return line
    
    def line_plot_y(self):
        data= self.rel_data()
        for dict in lines_dict_y:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Line_y"]
        return line
    
    def scatter_plot(self):
        data= self.rel_data()
        for dict in scatter_dict:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Scatter"]
        return line
    
    def pos_hist(self):
        data= self.rel_data()
        for dict in pos_dict:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Pos"]
        return line
    
    def neu_hist(self):
        data= self.rel_data()
        for dict in neu_dict:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Neu"]
        return line
    
    def neg_hist(self):
        data= self.rel_data()
        for dict in neg_dict:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Neg"]
        return line
    
    def comp_hist(self):
        data= self.rel_data()
        for dict in comp_dict:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Comp"]
        return line
    
    def pol_plot(self):
        data= self.rel_data()
        for dict in pol_dict:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Polp"]
        return line
    
    def vcomp_plot(self):
        data= self.rel_data()
        for dict in vcomp_dict:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Vcomp"]
        return line
    
    def lstm_plot(self):
        data= self.rel_data()
        for dict in lstm_dict:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Lstm"]
        return line
    
    def vpol_plot(self):
        data= self.rel_data()
        for dict in vpol_dict:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Vpol"]
        return line
    
    def vcompr_plot(self):
        data= self.rel_data()
        for dict in vcompr_dict:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Vcompr"]
        return line
    
    def rnn_plot(self):
        data= self.rel_data()
        for dict in rnn_dict:
            for k,v in dict.items():
                if v == self.Crypto:
                    line=dict["Rnn"]
        return line
```

```python
CD = CryptoDashboard()
```

```python
#Dashboard elements
dash_title = "# NLP-Crypto-Project"

#Dashboard description
dash_desc = " Our project aims to examine the relationship between public sentiment indicators and the prices of various cryptocurrencies. Sentiment is gathered from different news sources and run through VADER and LSTM"

#Placing plots in tabs
crypto_row = pn.Row(
    CD.param, CD.logo_png
)

heat_tabs = pn.Tabs(
    ("Daily", CD.heat_plot_d),
    ("Weekly", CD.heat_plot_w),
    ("Monthly", CD.heat_plot_m),
    ("Yearly", CD.heat_plot_y)
)

line_tabs = pn.Tabs(
    ("Daily", CD.line_plot_d),
    ("Weekly", CD.line_plot_w),
    ("Monthly", CD.line_plot_m),
    ("Yearly", CD.line_plot_y)
)

hist_tabs = pn.Tabs(
    ("Positive", CD.pos_hist), 
    ("Neutral", CD.neu_hist), 
    ("Negative", CD.neg_hist), 
    ("Compound", CD.comp_hist)
)

pol_row = pn.Row(
    CD.pol_plot, CD.vpol_plot
)

comp_row = pn.Row(
    CD.vcomp_plot, CD.vcompr_plot
)

lstm_rnn_row = pn.Row(
    CD.lstm_plot, CD.rnn_plot
)

cm_tabs = pn.Tabs(
    ("Polarity", pol_row),
    ("Compound", comp_row),
    ("LSTM-RNN", lstm_rnn_row)
)

#Dashboard construction
dashboard = pn.Column(
    dash_title,
    dash_desc,
    crypto_row,
    "## Heatmap",
    heat_tabs,
    "## Line Plot",
    line_tabs,
    "## Scatter Plot",
    CD.scatter_plot,
    "## Histograms",
    hist_tabs,
    cm_tabs
)
```
